```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import polars as pl
```

```python
TOOLS = "pan,wheel_zoom,box_zoom,reset,save,box_select"
output_notebook()
```

```python
N = 10000

x = np.random.normal(0, np.pi, N)
y = np.sin(x) + np.random.normal(0, 0.2, N)

p = figure(tools=TOOLS, output_backend="webgl")
p.scatter(x, y, alpha=0.1, nonselection_alpha=0.001)

show(p)
```

```python
raw_data = (
    pl.scan_csv(
        "~/Dokumente/Waveforms/freqGenTest-2000s.csv",
        skip_rows_after_header=2,
        separator=";",
        decimal_comma=True,
    )
).collect()

current_df = raw_data.select("Channel A")
current_np = current_df.to_numpy()
current_np = np.ravel(current_np)
time = np.linspace(0, (len(current_np) - 1) / 8000, len(current_np))
```

```python
p = figure(tools=TOOLS, output_backend="webgl")
p.line(time, current_np)
show(p)
```
